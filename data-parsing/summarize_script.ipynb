```python
import pandas as pd
import datetime
```

```python
File = "GFW-gear/trollers.csv"
df = pd.read_csv(File)
df.head()
```

```python
df.drop_duplicates(subset=['mmsi', 'lat', 'lon', 'is_fishing'], ignore_index=True, inplace=True)
df.head()
```

```python
df.drop(df[df['source'] == 'false_positives'].index, inplace = True)
len(df)
```

```python
# df['date'] = datetime.datetime.fromtimestamp(df['timestamp'].astype(int)).strftime('%Y-%m-%d')
df['date'] = df['timestamp'].apply(lambda t: datetime.datetime.utcfromtimestamp(t).strftime('%Y-%m-%d'))
df.head()
```

```python
df['source'].value_counts()
```

```python
df.drop('source', axis=1, inplace=True)
```

```python
df.drop('distance_from_shore', axis=1, inplace=True)
```

```python
df.drop('distance_from_port', axis=1, inplace=True)
```

```python
df.drop('timestamp', axis=1, inplace=True)
```

```python
df.head()
```

```python
yeardata = df.loc[(df['date'] >= '2017-01-01') & (df['date'] <= '2017-12-31')]
yeardata
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
yeardata['mmsi'].value_counts()
```

```python
mmsi2012 = len(yeardata['mmsi'].unique())
records2012 = len(yeardata['mmsi'])
print(f'2012 has {mmsi2012} unique vessels, {records2012} data points')
```

```python
yeardata.head()
```

```python
# Initialize dictionary
summary = {"Vessel Type": [], "Year": [], "Num of Vessels": [], "Data Points": []}
```

```python
# Update years and vessel type
years = [2012, 2013, 2014, 2015, 2016]
vtype = 'Trollers'

for year in years:
    start = str(year) + '-01-01'
    end = str(year) + '-12-31'
    yeardata = df.loc[(df['date'] >= start) & (df['date'] <= end)]
    vessels = len(yeardata['mmsi'].unique())
    records = len(yeardata['mmsi'])
    summary["Vessel Type"].append(vtype)
    summary["Year"].append(year)
    summary["Num of Vessels"].append(vessels)
    summary["Data Points"].append(records)
    
print(summary)
```

```python
summarydf = pd.DataFrame.from_dict(summary, orient='columns')
summarydf
```

```python
summarydf.to_json('../data/all_vessels_summary.json', orient='records')
```
