```python
# Dependencies
import pandas as pd
import json
```

```python
# Store filepath in a variable
capture_quantity_file="Capture_2021.1.2/CAPTURE_QUANTITY.csv"
# Read our Data file with the pandas library
capture_quantity_df=pd.read_csv(capture_quantity_file)

# Rename columns
capture_quantity_df2=capture_quantity_df.rename(columns={
    'COUNTRY.UN_CODE':'COUNTRY_UN_CODE',
    'SPECIES.ALPHA_3_CODE':'SPECIES_ALPHA_3_CODE',
    'AREA.CODE':'AREA_CODE'
})
# Replace NaN with 'Of'
capture_quantity_df2['STATUS']=capture_quantity_df2['STATUS'].fillna('Of')
# Display dataframe
capture_quantity_df2
```

```python
# Store filepath in a variable
unit_file="Capture_2021.1.2/FSJ_UNIT.csv"
# Read our Data file with the pandas library
unit_df=pd.read_csv(unit_file)
# Drop columns that are not needed
unit_df2=unit_df.drop([
    'Description_Es',
    'Description_Fr',
    'Name_Ar',
    'Name_Cn',
    'Name_Es',
    'Name_Fr',
    'Name_Ru'
],axis='columns')
# Rename columns
unit_df3=unit_df2.rename(columns={
    'Description_En':'MEASURE_DESCRIPTION',
    'Identifier':'MEASURE_IDENTIFIER',
    'Multiplier':'MEASURE_MULTIPLIER',
    'Name_En':'MEASURE_NAME',
    'SWS_Unit':'MEASURE_SWS_UNIT',
    'Unit':'MEASURE_UNIT'
})
# Display dataframe
unit_df3
```

```python
# Store filepath in a variable
country_groups_file="Capture_2021.1.2/CL_FI_COUNTRY_GROUPS.csv"
# Read our Data file with the pandas library
country_groups_df=pd.read_csv(country_groups_file)
# Set the title case for the "Official_Name_En" column
country_groups_df['Official_Name_En'] = country_groups_df['Official_Name_En'].str.title()
# Create a new dataframe with only the needed columns
country_groups_df2=country_groups_df[[
    'UN_Code',
    'Identifier',
    'ISO2_Code',
    'ISO3_Code',
    'Name_En',
    'Official_Name_En',
    'Continent_Group_En',
    'EcoClass_Group_En',
    'GeoRegion_Group_En'
]]
# Rename columns
country_groups_df3=country_groups_df2.rename(columns={
    'Identifier':'COUNTRY_IDENTIFIER',
    'ISO2_Code':'COUNTRY_ISO2_CODE',
    'ISO3_Code':'COUNTRY_ISO3_CODE',
    'Name_En':'COUNTRY_NAME',
    'Official_Name_En':'COUNTRY_OFFICIAL_NAME',
    'Continent_Group_En':'COUNTRY_CONTINENT',
    'EcoClass_Group_En':'COUNTRY_ECONOMIC_CLASS',
    'GeoRegion_Group_En':'COUNTRY_GEOGRAPHIC_REGION'
})
# Display dataframe
country_groups_df3
```

```python
# Store filepath in a variable
species_groups_file="Capture_2021.1.2/CL_FI_SPECIES_GROUPS.csv"
# Read our Data file with the pandas library
species_groups_df=pd.read_csv(species_groups_file)
# Create a new dataframe with only the needed columns
species_groups_df2=species_groups_df[[
    '3A_Code',
    'Taxonomic_Code',
    'Identifier',
    'Name_En',
    'Scientific_Name',
    'Author',
    'Major_Group_En',
    'Yearbook_Group_En',
    'ISSCAAP_Group_En',
    'CPC_Group_En'
]]
# Rename columns
species_groups_df3=species_groups_df2.rename(columns={
    'Taxonomic_Code':'SPECIES_TAXONOMIC_CODE',
    'Identifier':'SPECIES_IDENTIFIER',
    'Name_En':'SPECIES_NAME',
    'Scientific_Name':'SPECIES_SCIENTIFIC_NAME',
    'Major_Group_En':'SPECIES_MAJOR_GROUP',
    'Yearbook_Group_En':'SPECIES_YEARBOOK_GROUP',
    'ISSCAAP_Group_En':'SPECIES_ISSCAAP_GROUP',
    'CPC_Group_En':'SPECIES_CPC_GROUP'
})
# Display dataframe
species_groups_df3
```

```python
# Store filepath in a variable
water_area_groups_file="Capture_2021.1.2/CL_FI_WATERAREA_GROUPS.csv"
# Read our Data file with the pandas library
water_area_groups_df=pd.read_csv(water_area_groups_file)
# Create a new dataframe with only the needed columns
water_area_groups_df2=water_area_groups_df[[
    'Code',
    'Name_En',
    'Ocean_Group_En',
    'InlandMarine_Group_En',
    'FARegion_Group_En'
]]
# Rename columns
water_area_groups_df3=water_area_groups_df2.rename(columns={
    'Name_En':'AREA_NAME',
    'Ocean_Group_En':'AREA_OCEAN_GROUP',
    'InlandMarine_Group_En':'AREA_INLAND_MARINE_GROUP',
    'FARegion_Group_En':'AREA_FAR_REGION_GROUP'
})
# Display dataframe
water_area_groups_df3
```

```python
# Store filepath in a variable
symbol_file="GlobalProduction_2021.1.2/CL_FI_SYMBOL.csv"
# Read our Data file with the pandas library
symbol_df=pd.read_csv(symbol_file)
# Create a new dataframe with only the needed columns
symbol_df2=symbol_df[[
    'Alternate',
    'Identifier',
    'Name_En',
    'Symbol'
]]
# Rename columns
symbol_df3=symbol_df2.rename(columns={
    'Alternate':'STATUS_ALTERNATE',
    'Identifier':'STATUS_IDENTIFIER',
    'Name_En':'STATUS_NAME'
})
# Display dataframe
symbol_df3
# Replace [blank] with NaN
symbol_df4=symbol_df3.replace(to_replace ="[blank]",value ="Of")
symbol_df4
```

```python
# Create combined dataframe, combine first the capture quantity and country dataframes
combined_df = capture_quantity_df2.merge(country_groups_df3, left_on='COUNTRY_UN_CODE', right_on='UN_Code')
# Drop the "UN_Code" column because it contains redundant information
combined_df.drop('UN_Code', axis='columns', inplace=True)
# Display dataframe
combined_df.head()
```

```python
# Build the combined dataframe, add the species information
combined_df2 = combined_df.merge(species_groups_df3, left_on='SPECIES_ALPHA_3_CODE', right_on='3A_Code')
# Drop the "3A_Code" column because it contains redundant information
combined_df2.drop(['3A_Code','Author'], axis='columns', inplace=True)
# Display dataframe
combined_df2.head()
```

```python
# Build the combined dataframe, add the water area information
combined_df3 = combined_df2.merge(water_area_groups_df3, left_on='AREA_CODE', right_on='Code')
# Drop the "Code" column because it contains redundant information
combined_df3.drop('Code', axis='columns', inplace=True)
# Display dataframe
combined_df3.head()
```

```python
# Build the combined dataframe, add the units information
combined_df4 = combined_df3.merge(unit_df3, left_on='MEASURE', right_on='Code')
# Drop the "Code" column because it contains redundant information
combined_df4.drop('Code', axis='columns', inplace=True)
# Display dataframe
combined_df4.head()
```

```python
# Build the combined dataframe, add the production source information
combined_df5 = combined_df4.merge(symbol_df4, left_on='STATUS', right_on='Symbol')
# Drop the "Code" column because it contains redundant information
combined_df5.drop('Symbol', axis='columns', inplace=True)
# Display dataframe
combined_df5
```

```python
# Arrange columns in logical order
combined_df6 = combined_df5[[
    'COUNTRY_UN_CODE',
    'COUNTRY_IDENTIFIER',
    'COUNTRY_ISO2_CODE',
    'COUNTRY_ISO3_CODE',
    'COUNTRY_NAME',
    'COUNTRY_OFFICIAL_NAME',
    'COUNTRY_CONTINENT',
    'COUNTRY_ECONOMIC_CLASS',
    'COUNTRY_GEOGRAPHIC_REGION',
    'SPECIES_ALPHA_3_CODE',
    'SPECIES_TAXONOMIC_CODE',
    'SPECIES_IDENTIFIER',
    'SPECIES_NAME',
    'SPECIES_SCIENTIFIC_NAME',
    'SPECIES_MAJOR_GROUP',
    'SPECIES_YEARBOOK_GROUP',
    'SPECIES_ISSCAAP_GROUP',
    'SPECIES_CPC_GROUP',
    'AREA_CODE',
    'AREA_NAME',
    'AREA_OCEAN_GROUP',
    'AREA_INLAND_MARINE_GROUP',
    'AREA_FAR_REGION_GROUP',
    'PERIOD',
    'VALUE',
    'MEASURE',
    'MEASURE_DESCRIPTION',
    'MEASURE_IDENTIFIER',
    'MEASURE_MULTIPLIER',
    'MEASURE_NAME',
    'MEASURE_SWS_UNIT',
    'MEASURE_UNIT',
    'STATUS',
    'STATUS_ALTERNATE',
    'STATUS_IDENTIFIER',
    'STATUS_NAME'
]]
# Display dataframe
combined_df6
```

```python
# Create a clean DataFrame by selecting only "PISCES" from the "SPECIES.MAJOR_GROUP" column
pisces_only_df=combined_df6.loc[combined_df6['SPECIES_MAJOR_GROUP']=='PISCES', :]
# Display dataframe
pisces_only_df
```

```python
# Create a clean DataFrame by selecting only fish located in "Marine areas"
marine_pisces_only_df=pisces_only_df.loc[pisces_only_df['AREA_INLAND_MARINE_GROUP']=='Marine areas', :]
# Display dataframe
marine_pisces_only_df
```

```python
# Drop columns that are not needed
marine_pisces_only_df2=marine_pisces_only_df.drop([
    'COUNTRY_UN_CODE',
    'COUNTRY_IDENTIFIER',
    'COUNTRY_ISO2_CODE',
    'COUNTRY_ISO3_CODE',
#     'COUNTRY_NAME',
    'COUNTRY_OFFICIAL_NAME',
#     'COUNTRY_CONTINENT',
#     'COUNTRY_ECONOMIC_CLASS',
#     'COUNTRY_GEOGRAPHIC_REGION',
    'SPECIES_ALPHA_3_CODE',
    'SPECIES_TAXONOMIC_CODE',
    'SPECIES_IDENTIFIER',
    'SPECIES_NAME',
#     'SPECIES_SCIENTIFIC_NAME',
    'SPECIES_MAJOR_GROUP',
    'SPECIES_YEARBOOK_GROUP',
#     'SPECIES_ISSCAAP_GROUP',
#     'SPECIES_CPC_GROUP',
#     'AREA_CODE',
#     'AREA_NAME',
#     'AREA_OCEAN_GROUP',
    'AREA_INLAND_MARINE_GROUP',
    'AREA_FAR_REGION_GROUP',
#     'PERIOD',
#     'VALUE',
    'MEASURE',
    'MEASURE_DESCRIPTION',
    'MEASURE_IDENTIFIER',
    'MEASURE_MULTIPLIER',
#     'MEASURE_NAME',
    'MEASURE_SWS_UNIT',
#     'MEASURE_UNIT',
    'STATUS',
    'STATUS_ALTERNATE',
    'STATUS_IDENTIFIER',
#     'STATUS_NAME'
],axis='columns')
# Display dataframe
marine_pisces_only_df2
```

```python
# Replace NaN with ''
marine_pisces_only_df3=marine_pisces_only_df2.fillna('')
# Sort the data by 'COUNTRY_ISO2_CODE','SPECIES_ALPHA_3_CODE',and 'PERIOD' columns
marine_pisces_only_df4=marine_pisces_only_df3.sort_values(by=['COUNTRY_NAME','SPECIES_SCIENTIFIC_NAME','PERIOD'],ascending=[True,True,True])
# Filter for years 2010 or newer
marine_pisces_only_df5=marine_pisces_only_df4.loc[marine_pisces_only_df4['PERIOD']>=2010, :]
# Reset index values
marine_pisces_only_df5=marine_pisces_only_df5.reset_index(drop=True)
# Display dataframe
marine_pisces_only_df5
```

```python
# Group dataframe by two columns ("SPECIES_ISSCAAP_GROUP" & "PERIOD")
grouped_marine_pisces_only_data = marine_pisces_only_df5.groupby(['COUNTRY_NAME', 'SPECIES_ISSCAAP_GROUP', 'PERIOD'])
```

```python
# Converting a GroupBy object into a DataFrame
grouped_marine_pisces_only_df = pd.DataFrame(grouped_marine_pisces_only_data["VALUE"].sum())
# Display dataframe
grouped_marine_pisces_only_df
```

```python
# Reset index values
grouped_marine_pisces_only_df2=grouped_marine_pisces_only_df.reset_index(drop=False)
# Display dataframe
grouped_marine_pisces_only_df2
```

```python
# Store data in JSON format
grouped_marine_pisces_only_df2.to_json('data/numbers_captured_marine_fish.json',orient='records')
```
